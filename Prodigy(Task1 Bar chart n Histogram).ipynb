```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount("/content/drive")
```

```python


# Load the CSV file (update the path if necessary)
file_path = "/content/Cyber Safety Students Data Collection Sheet 2024-25 (1).csv"
df = pd.read_csv(file_path)

# Clean column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()

# Bar Chart: Gender Distribution
gender_column = 'Fender (M /F)'  # Correct column name after inspection
gender_counts = df[gender_column].value_counts()

# Histogram: Marks Distribution
marks = df['Marks'].dropna()

# Set up the plotting area
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot bar chart for gender
sns.barplot(x=gender_counts.index, y=gender_counts.values, ax=axes[0], palette='Set2')
axes[0].set_title('Gender Distribution')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Number of Students')

# Plot histogram for marks
sns.histplot(marks, bins=10, kde=True, ax=axes[1], color='skyblue')
axes[1].set_title('Marks Distribution')
axes[1].set_xlabel('Marks')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
